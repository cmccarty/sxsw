# Trip Advisor API Examples

Below are examples for all of the API endpoints made available through the Trip Advisor API. 

The core operations are:
- Search (/search)
- Nearby (/map)
- Location (/location)
- Reviews (/reviews)


Search allows you to search by keywords for matching locations (hotels, restaurants, attractions, or geos).  You can restrict searches to a particular geo (eg Austin), as well as certain category types (eg. restaurants).

Nearby allows you to specify a lat/lon pair to search near. You can set a radius if you like, and order by proximity or popularity. You can also add category types, a geo, or a search query to further filter.

/Location allows you to get more information about a location.  If you don't specify any connections, you will get meta data (name, address, num_reviews, ratings, etc).  You can also specify /attractions, /restaurants, /hotels to list other location types within a geo. You can get more results by paging with offset, limit.

## Map

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&limit=10' % (lat, lng, api_key)
response = urllib.urlopen(url).read()
nearby_restaurants = json.loads(response)
for restaurant in nearby_restaurants['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (restaurant['name'], restaurant['distance'], restaurant.get('rating'), restaurant['numReviews'], restaurant.get('cuisine'))
```

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&sort=popularity&distance=%f&limit=10' % (lat, lng, api_key, 0.25)
response = urllib.urlopen(url).read()
nearby_restaurants = json.loads(response)
for restaurant in nearby_restaurants['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (restaurant['name'], restaurant['distance'], restaurant.get('rating'), restaurant['numReviews'], restaurant.get('cuisine'))
```

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&limit=3&q=bar' % (lat, lng, api_key)
response = urllib.urlopen(url).read()
nearby_bars = json.loads(response)
for bar in nearby_bars['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (bar['name'], bar['distance'], bar.get('rating'), bar['numReviews'], bar.get('cuisine'))
```
