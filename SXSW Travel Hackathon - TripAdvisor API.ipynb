# Trip Advisor API Examples

<p>Below are examples for all of the API endpoints made available through the Trip Advisor API.</p>

<p>
The core operations are:<br/>
- <a href="#location">Location (/location)</a><br>
- <a href="#reviews">Reviews (/reviews)</a><br>
- <a href="#nearby">Nearby (/map)</a><br>
- <a href="#search">Search (/search)</a>
</p>

<p>Search allows you to search by keywords for matching locations (hotels, restaurants, attractions, or geos). You can restrict searches to a particular geo (eg Austin), as well as certain category types (eg. restaurants).</p>

<p>Nearby allows you to specify a lat/lon pair to search near. You can set a radius if you like, and order by proximity or popularity. You can also add category types, a geo, or a search query to further filter.</p>

<p>Location allows you to get more information about a location. If you don't specify any connections, you will get meta data (name, address, num_reviews, ratings, etc). You can also specify /attractions, /restaurants, /hotels to list other location types within a geo. You can get more results by paging with offset, limit.</p>

## Nearby

<p>You can also search for locations using a known lat/long pair. This is useful if you are looking for places nearby. For this we use /map instead of /location.</p>

<p>For example, here are the 10 closest restaurants to our location</p>

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&limit=10' % (lat, lng, api_key)
response = urllib.urlopen(url).read()
nearby_restaurants = json.loads(response)
for restaurant in nearby_restaurants['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (restaurant['name'], restaurant['distance'], restaurant.get('rating'), restaurant['numReviews'], restaurant.get('cuisine'))
```

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&sort=popularity&distance=%f&limit=10' % (lat, lng, api_key, 0.25)
response = urllib.urlopen(url).read()
nearby_restaurants = json.loads(response)
for restaurant in nearby_restaurants['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (restaurant['name'], restaurant['distance'], restaurant.get('rating'), restaurant['numReviews'], restaurant.get('cuisine'))
```

```python
lat = 30.26532
lng = -97.73852
url = 'http://api.tripadvisor.com/api/mobex/1.1/map/%f,%f/restaurants?key=%s&limit=3&q=bar' % (lat, lng, api_key)
response = urllib.urlopen(url).read()
nearby_bars = json.loads(response)
for bar in nearby_bars['items']:
    print '%s is %s mi away\n\trating: %s, reviews: %s\n\t%s\n' % (bar['name'], bar['distance'], bar.get('rating'), bar['numReviews'], bar.get('cuisine'))
```

## Search

```python
q = "museum"
url = 'http://api.tripadvisor.com/api/mobex/1.1/search/%s?key=%s&limit=20' % (q, api_key)
response = urllib.urlopen(url).read()
#print response
```

```python
# Note that these are grouped by category type within the 'items' element.
json_obj = json.loads(response)
print json_obj['items'].keys()
```

```python
locations = json.loads(response)
for category in locations['items']:
    print '%s (%d)' % (category, len(locations['items'][category]))
    for loc in locations['items'][category]:
        print '\t%s (rating: %s, reviews: %s)' % (loc['name'], loc.get('rating'), loc['numReviews'])
```

```python
q = "capitol"
geo_id = 30196 # Austin, TX

url = 'http://api.tripadvisor.com/api/mobex/1.1/search/%s?key=%s&geo=%d&category=attractions' % (q, api_key, geo_id)
response = urllib.urlopen(url).read()
locations = json.loads(response)
for category in locations['items']:
    print '%s (%d)' % (category, len(locations['items'][category]))
    for loc in locations['items'][category]:
        print '\t%s (rating: %s, reviews: %s)' % (loc['name'], loc.get('rating'), loc['numReviews'])
```

More information and API Keys available at <a href="https://github.com/cmccarty/sxsw">github.com/cmccarty/sxsw</a>

```python
def get_num_attractions(location_id):
    url = 'http://api.tripadvisor.com/api/mobex/1.1/location/%d?key=%s&category=geos' % (location_id, api_key)
    response = urllib.urlopen(url).read()
    location = json.loads(response)
    if not location.get('items'): return None
    
    return location['items']['attractionCount']
    

def get_stats(city):
    # pull city
    url = 'http://api.tripadvisor.com/api/mobex/1.1/search/%s?key=%s&category=geos' % (city, api_key)
    response = urllib.urlopen(url).read()
    geos = json.loads(response)
    
    if not 'items' in geos: return None
    
    if geos['items'].get('geo'):
        top_geo = geos['items']['geo'][0]
        
        num_attractions = get_num_attractions(top_geo['locationId'])
      
        stats = {'id': top_geo['locationId'], 'name': top_geo['name'], 'num_attractions': num_attractions}
        return stats

    return None

# Find the cities with the most attractions
cities = ['San Francisco','Austin','Chicago','Boston','Paris','London','Rome']
stats = [get_stats(c) for c in cities]
stats = [s for s in stats if s] # filter out cities we couldn't find

stats = sorted(stats, key=lambda k: k['num_attractions'], reverse=True)

print '\n'.join([str(s) for s in stats if s])
    
    
    
    
```
